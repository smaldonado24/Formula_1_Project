```python
# Import packaged and plug-ins
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
from datetime import datetime

%matplotlib inline

# Assign a path for each CSV File
drivers_path = "./csv_files/drivers.csv"
lap_times_path = "./csv_files/lap_times.csv"
results_path = "./csv_files/results.csv"
constructor_standings_path = "./csv_files/constructor_standings.csv"
constructors_path = "./csv_files/constructors.csv"
driver_standings_path = "./csv_files/driver_standings.csv"
circuits_path = "./csv_files/circuits.csv"
constructor_results_path = "./csv_files/constructor_results.csv"
pit_stops_path = "./csv_files/pit_stops.csv"
qualifying_path = "./csv_files/qualifying.csv"
races_path = "./csv_files/races.csv"
seasons_path = "./csv_files/seasons.csv"
sprint_results_path = "./csv_files/sprint_results.csv"
status_path = "./csv_files/status.csv"

# Create dataframes for each CSV path
drivers_df = pd.read_csv(drivers_path)
lap_times_df = pd.read_csv(lap_times_path)
results_df = pd.read_csv(results_path)
constructors_standings_df = pd.read_csv(constructor_standings_path)
constructors_df = pd.read_csv(constructors_path)
driver_standings_df = pd.read_csv(driver_standings_path)
circuits_df = pd.read_csv(circuits_path)
constructor_results_df = pd.read_csv(constructor_results_path)
pit_stops_df = pd.read_csv(pit_stops_path)
qualifying_df = pd.read_csv(qualifying_path)
races_df = pd.read_csv(races_path)
seasons_df = pd.read_csv(seasons_path)
sprint_results_df = pd.read_csv(sprint_results_path)
status_df = pd.read_csv(status_path)
```

```python
# Merge "Results" and "Races" CSVs into a single dataframe
results_races_df_1 = pd.merge(results_df, races_df, on = 'raceId', how='left')

# Delete unneeded data columns to clean dataframe 
results_races_df_1.drop(['number', 'position', 'positionText', 'laps', 'points', 'milliseconds', 'fastestLapTime', 'fastestLapSpeed', 'fp1_date', 'time_y',
           'fastestLap', 'statusId', 'resultId', 'time_x', 'fp1_time', 'fp2_time', 'fp3_time', 'fp2_date', 'fp3_date', 'quali_date', 'url',
           'raceId', 'quali_time', 'sprint_date', 'sprint_time'], axis=1, inplace=True)
results_races_df_1
```

```python
results_races_drivers_df_2 = pd.merge(results_races_df_1, drivers_df, on = 'driverId', how='left')
results_races_drivers_df_2.drop(['date', 'url', 'dob', 'rank'], axis=1, inplace=True)
results_races_drivers_df_2
```

```python
complete_merged_df = pd.merge(results_races_drivers_df_2, constructors_df, on = 'constructorId', how= 'left')
complete_merged_df.drop(['number', 'url', 'constructorRef', 'driverRef', 'grid'], axis=1, inplace=True)
complete_merged_df
```

```python
# Combine columns "forename" and "surname" into single column "full_name" 
complete_merged_df['full_name'] = complete_merged_df[['forename', 'surname']].agg(' '.join, axis=1)
complete_merged_df.drop(['forename', 'surname'], axis=1, inplace=True)
complete_merged_df
```

```python
# Rename default column names to clearly identify 
complete_merged_df.rename(columns={'name_x':'gp_name', 'nationality_x':'driver_nationality',
                     'name_y':'constructor_name', 'nationality_x':'driver_nationality', 
                     'nationality_y':'constructor_nationality'},inplace = True)
complete_merged_df
```

```python
#Print info to check datatypes of combined dataframes
complete_merged_df.info()
```

```python
# Setting plots sizes
#sb.set_palette('Set3')
#plt.rcParams['figure.figsize']=10,6
```

```python
# Created a new dataframe to show total for all-time Grand Prix winners by position order equaling to first place, grouping by the drivers full name, in an ascending order.
grand_prix_winner = complete_merged_df.loc[complete_merged_df['positionOrder']==1].groupby('full_name')['positionOrder'].count().sort_values(ascending=False).to_frame().reset_index()
grand_prix_winner
```

```python
# Create a bar-plot chart of all Grand Prix winners
sb.barplot(data=grand_prix_winner, y='full_name', x='positionOrder',color='red', alpha=0.8)
plt.title('Most Grand Prix winners in F1')
plt.ylabel('Driver Name')
plt.xlabel('Number of GP wins')
#plt.yticks([])
```

```python
# Create a new dataframe to show the top 10 all-time Grand Prix winners
top10drivers = grand_prix_winner.head(10)
print(top10drivers)
```

```python
# Create a bar-plot chart that shows top 10 all-time Grand Prix winners
sb.barplot(data=top10drivers, y='full_name', x='positionOrder',color='red', alpha=0.8, linewidth=.8, edgecolor='black')
plt.title('Most Grand Prix winners in F1')
plt.ylabel('Driver Name')
plt.xlabel('Number of GP wins')
```

```python
# Create a datafrane that shows the top Grand Prix constructor winners
constructor_wins = complete_merged_df.loc[complete_merged_df['positionOrder']==1].groupby('constructor_name')['positionOrder'].count().sort_values(ascending=False).to_frame().reset_index()
constructor_wins.head(10)
```

```python
#Create a bar-plot of all Grand Prix constructor winners
sb.barplot(data=constructor_wins, y='constructor_name', x='positionOrder',color='blue', alpha=0.8)
plt.title('Most Grand Prix Constructor Wins in F1')
plt.ylabel('Constructor Name')
plt.xlabel('Number of GP wins')
#plt.yticks([])
```

```python
# Create a new dataframe to show the top 10 Grand Prix Constructor wins
top10constructors = constructor_wins.head(10)
print(top10constructors)
```

```python
# Create a bar-plot chart that shows top 10 all-time constructor Grand Prix winners
sb.barplot(data=top10constructors, y='constructor_name', x='positionOrder',color='blue', alpha=0.8, linewidth=.8, edgecolor='black')
plt.title('Most Grand Prix Constructor Wins in F1')
plt.ylabel('Constructor Name')
plt.xlabel('Number of GP wins')
```

```python
###Create a chart that shows who is the all time winner at the Monaco Grand Prix###
#Starting point to filter all time Monaco Grand Prix winner
filtered_df = complete_merged_df[complete_merged_df['gp_name'] == 'Monaco Grand Prix']
filtered_df
```

```python
#Create a filtered dataframe to drop unneeded columns
filtered_df.drop(['constructor_name', 'driver_nationality','constructor_nationality'], axis=1, inplace=True)
filtered_df
```

```python
# Created a new dataframe to show total for all-time Monaco Grand Prix winner by position order equaling to first place, grouping by the drivers full name, in an ascending order.
monaco_all_time_df = filtered_df.loc[filtered_df['positionOrder']==1].groupby('full_name')['positionOrder'].count().sort_values(ascending=False).to_frame().reset_index()
monaco_all_time_df
```

```python
#Create a bar-plot chart of the most Monaco Grand Prix wins
sb.barplot(data=monaco_all_time_df, y='full_name', x='positionOrder',color='gold', alpha=0.8, linewidth=.8, edgecolor='black')
plt.title('Most Monaco Grand Prix Winners')
plt.ylabel('Driver Name')
plt.xlabel('Number of GP wins')
```

```python
#Filter to show top 5 of all time
monaco_top_5_df = monaco_all_time_df.head(5)
print(monaco_top_5_df)
```
